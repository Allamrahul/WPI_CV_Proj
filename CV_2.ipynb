```python

def video_to_frames(video_path, save_path):
    if not os.path.isdir(save_path):
        os.mkdir(save_path)

    if len(os.listdir(save_path)) == 0:
        vidcap = cv2.VideoCapture(video_path)
        success,image = vidcap.read()
        count = 0

        while success:
            path = os.path.join(save_path , 'frame'+str(count)+'.jpg')
            cv2.imwrite(path, image)     # save frame as JPEG file
            success,image = vidcap.read()
            count += 1
    else:
        print("frames have already been extracted")
```

```python
from os.path import isfile, join
def convert_frames_to_video(pathIn,pathOut,fps):
    frame_array = []
    files = [f for f in os.listdir(pathIn) if isfile(join(pathIn, f))]
    print(files)
    #for sorting the file names properly
    #files.sort(key = lambda x: int(x[-1]))
    files.sort(key=lambda x: int(''.join(filter(str.isdigit, x))))
    print(files)
    for i in range(len(files)):
        filename=os.path.join(pathIn, files[i])
        print(filename)
        #reading each files
        img = cv2.imread(filename)
        
        print("IN")
        height, width, layers = img.shape
        size = (width,height)
        print(filename)
        #inserting the frames into an image array
        frame_array.append(img)
    out = cv2.VideoWriter(pathOut,cv2.VideoWriter_fourcc(*'DIVX'), fps, size)
    for i in range(len(frame_array)):
        # writing to a image array
        out.write(frame_array[i])
    out.release()
```
